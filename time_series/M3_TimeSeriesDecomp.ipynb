# Program Imports #

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
plt.style.use('seaborn')
```

```python
# Functions to use for time-series decomposition:
def baseplot(x, y):

    return plt.plot(x, y, figsize=(15, 6))

date_data = {'Jan' : ('01', '31'), 'Feb' : ('02', '28'), 'Mar' : ('03', '31'), 'Apr' : ('04', '30'), 'May' : ('05', '31'), 'Jun' : ('06', '30'), 'Jul' : ('07', '31'), 'Aug' : ('08', '31'), 'Sep' : ('09', '30'), 'Oct' : ('10', '31'), 'Nov' : ('11', '30'), 'Dec' : ('12', '31')}

def to_data_range(date_str):
    return '20' + date_str[-2:] + '-' + date_data[date_str[:3]][0] + '-' + date_data[date_str[:3]][0]

def correl(per_max=36, series=[]):

    maxCorr = 0.0
    period = np.nan
    for i in range(1, per_max):
        corr = series.autocorr(lag=i)
        if corr > maxCorr:
            maxCorr = corr
            period = i
    return {
        "Period": period,
        "Max_Correlation": maxCorr
    }
```

# Time Series Decomposition # 

This data was provided via the CTBA zip file in the course downloadables. The file used in this will be available in the repository of this project. The data is of the total amount/demand of/for Construction done nationally in monthly increments from 2002 to 2013. For expediency, I am dividing this analysis into 2 separate sections. One for Private Construction and one for Public Construction.

## Data Processing ##

First, the data needs to be prepared. Adding a datetime index will also be helpful since the data is time-series. 

```python
ts_data = pd.read_csv("ConstructionTimeSeriesDataV2.csv", header=0, sep=',')
```

```python
ts_data['Date'] = ts_data['Month-Year'].apply(to_data_range)
ts_data.index = pd.date_range(start=ts_data['Date'].iloc[0], periods = ts_data['Date'].shape[0], freq='M')
```

```python
 ts_data.head()
```

```python
# Split data into private and public construction.
private_construction = ts_data[['Private Construction']]
private_construction.columns = ['private_construction']
public_construction = ts_data[['Public Construction']]
public_construction.columns = ['public_construction']
```

## Private Construction ##

This is a time series decomposition of a time series of private construction in the United States from January 2002 to December 2013. The data will likely not have a linear trend, and a moving average method will likely need to be applied in order to properly decompose the time series data.

### Trend Analysis ###

The trend for this data does not appear to be linear in nature. Therefore linear regression will likely not be an effective method for this decomposition.

```python
private_construction.plot(linewidth=3, figsize=(15, 10))
```

```python
private_construction['m_avg'] = private_construction.private_construction.rolling(2).mean() # Conduct a rolling(moving) average of every 2 periods, omitting the 1st and last period.
private_construction.dropna(how="any", inplace=True)
```

```python
private_construction.plot(linewidth=3, figsize=(15,10))
```

```python
private_construction['sq_err'] = (private_construction['private_construction']-private_construction['m_avg'])**2
```

```python
sum(private_construction['sq_err'])/len(private_construction)
```

```python
private_construction['r1'] = private_construction['private_construction'] - private_construction['m_avg']
private_construction['r1_err'] = abs((private_construction['private_construction']-private_construction['r1'])/private_construction['private_construction'])
```

```python
private_construction['r1'].plot(figsize=(12,6))
print("Average Residual: ", private_construction['r1'].mean())
```

Based on the above graph, the residuals do not look to be correlated which means that the moving average with 3 periods is an acceptable model to use with this data.

### Cyclical Analysis - Private Construction ###

```python
correl(per_max=36, series=private_construction['r1'])
```

The above correlation would suggest the presence of a strong annual cycle in the data.

```python
period = 12
cycle = period
n_cycles = int(len(private_construction)/cycle + 0.5)
cycles = [private_construction.iloc[range(i*period, min((i+1)*period, len(private_construction)))]['r1'] for i in range(n_cycles)]
pts_cycles = [private_construction.iloc[range(i, len(private_construction['r1']), period)]['r1'].tolist() for i in range(period)]
avg = [sum(pts)/len(pts) for pts in pts_cycles]
```

```python
cycle = 12 #The cycle correlation that was found in the previous step.
n_cycle = int(len(private_construction)/cycle + 0.5) # Find the number of cycles which would be present in the data.
private_construction['cycle'] = np.nan
for i in range(len(private_construction)):
    private_construction.loc[private_construction.index[i], 'cycle'] = avg[i % cycle]
```

```python
fig, ax = plt.subplots()
ax.plot(private_construction['cycle'], label="Cyclical Pattern")
ax.plot(private_construction['r1'], label="Remainder After Trend")
ax.set_xlabel('Year')
ax.set_ylabel('Private Construction')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
fig.legend()
```

```python
private_construction['r2_private'] = private_construction['r1'] - private_construction['cycle']
private_construction['r2_err_private'] = abs(private_construction['r2_private']/private_construction['cycle'])
private_construction['model_fit'] = private_construction['m_avg'] + private_construction['cycle']
```

```python
print("Mean Absolute Percent Error: ", private_construction['r2_err_private'].mean())
```

```python
fig, ax = plt.subplots()
ax.plot(private_construction['private_construction'], label='Private Construction')
ax.plot(private_construction['model_fit'], label="Fit")
ax.set_xlabel("Year")
ax.set_ylabel("Private Construction")
fig.legend()
```

### Conclusions - Private Construction ###

There are definite limitations to the moving average approach to this dataset, specifically because the trend in the dataset is heavily influenced by periods that are more recent. This means that the larger window which was attempted made the performance of the model worse. I did try an exponential moving average, and the alpha level which worked best was around a=0.75 which would indicate the trend in the data is volatile and recency has a higher weight.

Aside from trend, there is a clear, precise annual trend to this data, with peaks in the summer months in the US (May, June, July, August) and troughs in the winter (November, December, January, Febuary) with ramp ups and ramp downs in the months adjacent to these. This is likely explained by weather and condition favorability of the Summer months with that of the winter.
